# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

years = [2008, 2008]
prices = [400, 380]
products = ['A', 'B']

plt.figure(figsize=(6, 4))
plt.scatter(years, prices, color='navy', s=28)  # Smaller dots

for i, label in enumerate(products):
    plt.text(years[i] + 0.07, prices[i], label, va='center', fontsize=8)  # Smaller font

plt.xlim(2007.5, 2014)
plt.ylim(0, 500)
plt.yticks(range(0, 501, 100), [f'${y}' for y in range(0, 501, 100)])
plt.xticks(range(2008, 2015))

plt.xlabel('Year')
plt.ylabel('Average price')
plt.title('Retail price over time', loc='left')

plt.suptitle('Products A and B were launched in 2008 at price points of $360+', 
             x=0.0, ha='left', fontsize=12, color='navy')
plt.subplots_adjust(top=0.85)

for spine in ['top', 'right']:
    plt.gca().spines[spine].set_visible(False)

plt.show()
```
