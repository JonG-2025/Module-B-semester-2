# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

months = [
"January","February","March","April","May","June",
"July","August","September","October","November","December"
]

ticket_received = [160, 180, 235, 145, 180, 160, 130, 200, 160, 135, 145, 175]
ticket_processed = [160, 180, 240, 145, 180, 150, 120, 155, 125, 100, 120, 140]

marker_start = 7

x = np.arange(len(months))

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, ticket_received, linestyle='-', color='grey', linewidth=2, label='Ticket Volume Received')
ax.plot(x, ticket_processed, linestyle='-', color='blue', linewidth=2, label='Ticket Volume Processed')

ax.plot(x[marker_start:], np.array(ticket_received)[marker_start:], marker='D', linestyle='None',
color='grey', markersize=7)
ax.plot(x[marker_start:], np.array(ticket_processed)[marker_start:], marker='s', linestyle='None',
color='blue', markersize=7)

offset = 6 
for xi, val in zip(x[marker_start:], np.array(ticket_received)[marker_start:]):
    ax.text(xi, val + 4, str(val), ha='center', va='bottom', color='grey', fontsize=9)
for xi, val in zip(x[marker_start:], np.array(ticket_processed)[marker_start:]):
    ax.text(xi, val + 4, str(val), ha='center', va='bottom', color='blue', fontsize=9)

ax.set_ylim(0, 300)
ax.set_xticks(x)
ax.set_xticklabels(months, rotation=45)
ax.grid(axis='y', linestyle='-', alpha=0.5)
ax.set_ylabel("Tickets")
ax.set_title("Figure 4.14 Page 115")
ax.legend()

plt.tight_layout()
plt.show()
```
