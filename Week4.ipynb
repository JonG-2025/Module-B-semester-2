```python
import numpy as np
import pandas as pd

np.random.seed(0)
num_data = 100
data = np.exp(np.random.uniform(size = num_data) * 4)
df = pd.DataFrame(data.T, columns = ["data"])
```

```python
print("The 75th percentile is:", df.quantile(q = 0.75))
df.plot.hist()
```

$Answer$: This is great for comparing prices across the board, however; it doesnt provide any specific <br>
catagories of drugs for comparison to their own prices since it is a random seed. Considering that, <br> it is labeled clearly for how often a number was generated for comparison to eachother. <br>
However, its a little harder to determine where the average of the prices are without further <br>
calculation and analysis. 

```python
df.plot.box()
```

$Answer$: This boxplot works well to address some of the short comings from the histogram. <br>
It allows for easy view of where most of the data is concentrated on as well as easily identifying outliers. <br> It also has a line where the 50th percentile falls right on the boxplot as opposed <br>
to being seperated from the visual like the histogram.

```python
# importing CSV file from data set into a pandas dataframe.

df = pd.read_csv("/workspaces/Module-B-semester-2/credit_risk_dataset.csv", sep=",")
```

```python
#Checking is all columns have a consistent count of data.

df.info()
```

```python
#Removing any rows with blank/NA fields.

df_clean = df.dropna()
```

```python
#Rechecking for consistent count across all columns.

df_clean.info()
```

```python
df_clean.describe()
```

```python
#Creating histograms for each column that is an integer or float now that data is cleaned.

df_clean.hist(figsize=(12, 8), bins=10, alpha=0.9)
```

```python
# Mean is above in df_cleaned.describe(). Id like to get the median and mode for loan amount.

s = df_clean['loan_amnt']

median = s.median()
mode_series = s.mode()

print("Loan Amount Median:", median)
print("Loan Amount Mode:", mode_series)
```

```python
#Creating a scatter plot to look at loan amount by age.

import matplotlib.pyplot as plt

x = df_clean['person_age']
y = df_clean.loc[x.index, "loan_amnt"]

data = df_clean[['person_age', 'loan_amnt']]
plt.figure(figsize=(10,6))
plt.scatter(data['person_age'], data['loan_amnt'], alpha=0.6, s=20)
plt.xlabel('Person Age')
plt.ylabel('Loan Amount')
plt.title('Loan Amount vs Person Age')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
#Creating a pie chart that shows loan interest rates.

rates = df_clean['loan_int_rate'].dropna()

bins = 4
labels = None
cats = pd.cut(rates, bins=bins)
counts = cats.value_counts().sort_index()

plt.figure(figsize=(7,7))
ax = counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, counterclock=False, ylabel='')
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.title('Loan Interest Rate Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
